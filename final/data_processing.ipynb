```python
import pandas as pd
import numpy as np
# from fancyimpute import KNN
pd.set_option('display.max_columns', 500)
```

```python
#load data
df=pd.read_csv('df.csv')
df=df.set_index('Customer ID')
```

```python
df
```

```python
#mapping
df=df.applymap(lambda x:1 if x=='Yes' else 0 if x=='No' else x)
df=df.applymap(lambda x:1 if x=='Male' else 0 if x=='Female' else x)
#frequency encoding
df['Zip Code']=df['Zip Code'].map(df['Zip Code'].value_counts().to_dict())
#df['Offer']=df['Offer'].map(df['Offer'].value_counts().to_dict())
# df['Internet Type']=df['Internet Type'].map(df['Internet Type'].value_counts().to_dict())
# df['Contract']=df['Contract'].map(df['Contract'].value_counts().to_dict())
# df['Payment Method']=df['Payment Method'].map(df['Payment Method'].value_counts().to_dict())
# df
#One hot encoding
df_offer_hot=pd.get_dummies(df['Offer']).drop('None',axis=1)
for column in df_offer_hot.columns:
    df[column]=df_offer_hot[column]
df=df.drop('Offer',axis=1)

df_internet_hot=pd.get_dummies(df['Internet Type']).drop('None',axis=1)
for column in df_internet_hot.columns:
    df[column]=df_internet_hot[column]
df=df.drop('Internet Type',axis=1)

df_contract_hot=pd.get_dummies(df['Contract'])
for column in df_contract_hot.columns:
    df[column]=df_contract_hot[column]
df=df.drop('Contract',axis=1)

df_payment_hot=pd.get_dummies(df['Payment Method'])
for column in df_payment_hot.columns:
    df[column]=df_payment_hot[column]
df=df.drop('Payment Method',axis=1)

# df
```

```python
df_nan_map=df.applymap(lambda x:1 if pd.isna(x) else 0)
df_nan_map
```
